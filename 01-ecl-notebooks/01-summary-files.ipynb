### 1. Import the needed stuff

import numpy as np
import pandas as pd

%matplotlib inline
from matplotlib import pyplot as plt

from ecl.summary import EclSum

### 2. Load and explore a sample case

```python
case = EclSum(r"C:\GitHub\opm-tests\norne\ECL.2014.2\NORNE_ATW2013.SMSPEC")
print(case)
```

```python
# Wells
print(case.wells())
```

```python
# Groups
print(case.groups())
```

```python
# Number of timesteps, e.g. the height of the table
print(len(case.get_dates()))
```

```python
# Number of columns, e.g. the width of the table
print(len(case.keys()))
```

### 3. Export to a Pandas Dataframe

```python
# Export the table of values to pandas dataframe
df = case.pandas_frame()
```

```python
df
```

You can notice that number of rows is identical to the number of timesteps. But the table has fewer columns compared to original `width`. This is because Eclipse has some extra columns, that we do not need. Basically, not all of them are real data from simulation.

### 4. Fixing the dataframe header

We can see that the header of the above table has old-ish, Eclipse Office-style format of object:vector:region. It is inconvenient to work with, e.g. it is not easy so subset some parameter or object.

Let's convert it to the Pandas' multiindex form. Attention: This is Pandas-specific, and will not work on other implementations of DataFrame (R, Julia, etc)

```python
v = list(map(lambda x: x.split(":"), df.columns))
col_header = pd.MultiIndex.from_arrays(pd.DataFrame(v).T.replace(np.nan, "").values)
df.columns = col_header
df
```

This format should look familiar to you, if you ever exported data from Petrel to Excel. The only thing we miss here is the units of measure, but exported dataframe did not contain it.

What is the advantage of such multi-indexed columns? See the next section.

### 5. Subsetting the data

Well, lets select a single parameter, say `WBHP`. We will use `head()` to get first 5 rows. Remember, we have 884 rows in total.

```python
# First method
df["WBHP"].head()
```

```python
# Second method. `axis=1` means columns. 0 means rows.
df.xs("WBHP", axis=1).head()
```

`df.xs()` is useful where you need to subset by second level of hierarchy, e.g. by well name.

```python
df.xs('D-1H', axis=1, level=1).head()
```

We can see that well `D-1H` has 51 parameters.

### 6. Plotting the data

No Jupyter Notebook is good without at least couple plots. Let's plot `WBHP` for several wells at once.

```python
df.xs("WBHP", axis=1).iloc[:,:9].plot(subplots=True, layout=(3,3), figsize=(12, 8))
```

### 7. Saving the data

You extracted, parsed, cleaned the data. You even plotted it. Good job! But the job is not done until data is saved.
Since 5000 columns are way over the Excel's limit, let's save a `WBHP` vector now. Notice that the `WBHP` word itself is in double brackets - this way the topmost header level (`WBHP` itself) is preserved.

```python
df[["WBHP"]].to_excel("norne-wbhp.xlsx")
```
